```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV
df = pd.read_csv('/content/Book1 (1).csv')
# Show basic structure
print("Dataset preview:")
print(df.head())

# Show column names
print("\nColumn names:")
print(df.columns)

# Show data types and missing values
print("\nInfo:")
print(df.info())

# Descriptive stats
print("\nSummary statistics:")
print(df.describe())

# Check for missing data
print("\nMissing values per column:")
print(df.isnull().sum())

# Drop rows with any missing values (optional)
df_cleaned = df.dropna()

# Plotting numeric columns (adjust based on your columns)
numeric_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns

# Histogram for numeric columns
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df_cleaned[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_cleaned[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV
df = pd.read_csv('Book1.csv')

# Show basic structure
print("Dataset preview:")
print(df.head())

# Show column names
print("\nColumn names:")
print(df.columns)

# Show data types and missing values
print("\nInfo:")
print(df.info())

# Descriptive stats
print("\nSummary statistics:")
print(df.describe())

# Check for missing data
print("\nMissing values per column:")
print(df.isnull().sum())

# Drop rows with any missing values (optional)
df_cleaned = df.dropna()

# Plotting numeric columns (adjust based on your columns)
numeric_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns

# Histogram for numeric columns
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df_cleaned[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_cleaned[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Save cleaned data
df_cleaned.to_csv('Book1_cleaned.csv', index=False)
print("\nCleaned data saved to Book1_cleaned.csv")

# Load CSV
df = pd.read_csv('Book1.csv')

# Show basic structure
print("Dataset preview:")
print(df.head())

# Show column names
print("\nColumn names:")
print(df.columns)

# Show data types and missing values
print("\nInfo:")
print(df.info())

# Descriptive stats
print("\nSummary statistics:")
print(df.describe())

# Check for missing data
print("\nMissing values per column:")
print(df.isnull().sum())

# Drop rows with any missing values (optional)
df_cleaned = df.dropna()

# Plotting numeric columns (adjust based on your columns)
numeric_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns

# Histogram for numeric columns
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df_cleaned[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_cleaned[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Save cleaned data
df_cleaned.to_csv('Book1_cleaned.csv', index=False)
print("\nCleaned data saved to Book1_cleaned.csv")
```
